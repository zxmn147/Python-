### 1. [簡答題] 請問下列兩種將 Array 轉換成 List 的方式有何不同？

```
print('list(a): ', list(a))
print('tolist(): ', a.tolist())
```


````python
import numpy as np
a = np.random.randint(10, size=6)
print('list(a): ', list(a)) 
print('tolist(): ', a.tolist()) 
print("----------------------------------")
b = np.random.randint(10, size=(3,4)) 
print('list(b): ', list(b))
print('tolist(): ', b.tolist())
print("----------------------------------")
c = np.random.randint(10, size=(2,3,2)) 
print('list(c): ', list(c))
print('tolist(): ', c.tolist())
print("----------------------------------")

#  由觀察得知
#　list(): 只會把第一層的元素轉換成List
#　tolist():全部資料的型態一起轉換
````

````python

a = np.random.randint(10, size=6) 
# 陣列維度
print('ndim: ', a.ndim)
# 維度大小
print('shape: ', a.shape)
# 陣列元素
print('size: ', a.size)
# 陣列資料型態
print('dtype: ', a.dtype)
# int32佔4個字節 / int64佔8個字節
print('itemsize: ', a.itemsize) 
# 資料長度
print('length: ', len(a))
# 資料型態
print('type: ', type(a))
````

````python
## 簡單版
def tolist(iterable):
    if type(iterable) != np.ndarray:
        return iterable
    newList = []
    for i in iterable:
        newList.append(tolist(i))
    return list(newList)

print(tolist(a))
print(tolist(b))
print(tolist(c))
````

````python
## 進階版
def tolist(iterable):
    if type(iterable) != np.ndarray:
        return iterable
    return [tolist(i) for i in iterable]
print(tolist(a))
print(tolist(b))
print(tolist(c))
````

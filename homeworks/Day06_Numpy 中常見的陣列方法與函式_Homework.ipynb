### 2. 請對一個 5x5 的隨機矩陣作正規化的操作。


````python
# 記得先 Import 正確的套件
import numpy as np
````

````python

# 正規化: 轉換0~1之間
A = np.random.random((5, 5))

A = (A - A.min()) / (A.max() - A.min())
A
````

### 3. 請建立一個長度等於 10 的正整數向量，並且將其中的最大值改成 -1。


````python
A = np.random.random((10))
print("A", A)
A[np.argmax(A)] = -1
A
````
